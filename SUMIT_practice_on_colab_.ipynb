```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
# NUMPY array can hold only one datatype at a time.
arr = np.array([1,2,3,4,5.5])
print(arr)
print(type(arr))
print(arr.dtype)
```

```python
# FEW FUNCTIONALITIES OF NUMPY.
arr = np.arange(10)
print(arr)
```

```python
#lets try decimal value gaps.
arr = np.arange(0,10,1.5)#start = 0,end = 10,step =1.5
print(arr)
```

```python
# MATRIX
arr = np.arange(12).reshape(3,4)
print(arr)
print(arr.shape)
```

```python
# LETS TRY ZEROS AND ONES IN NUMPY.
#ZEROS
arr = np.zeros(shape=(4,3))
print(arr)
```

```python
# ONES
arr = np.ones(shape=(4,4))
print(arr)
#THATS IT NUMPY CAN ONLY DO THIS WITH ZEROS AND ONES.
```

```python
arr = np.full((4,3),10)
print(arr)
```

```python
#ARTHEMATIC OPERATIONS IN NUMPY.
arr = np.array([1,2,3,4,5])
print(arr)

print(arr+2)
print(arr-2)
print(arr*2)
print(arr/2)
```

```python
# MEAN
arr = np.array([1,2,3,4,5,6,7])
print(arr)
print(np.mean(arr))
print(np.median(arr))
```

```python
arr = np.array([1,2,3,4,5,6,7])
print(arr)
print(arr>3)#IT WILL SHOW BOOLEAN .
print(arr[arr>3])# IT SHOWS NUMBERS GREATER THAN 3 ONLY.
```

```python
import numpy as np
#LETS TRY ARETHEMATIC OPERATIONS ON TWO ARRAYS SIMULTANEOUSLY.
arr1 = np.array([1,2,3,4,5,6])
arr2 = np.array([10,20,30,40,50,60])
print(arr1)
print(arr2)
#ADDING

print(arr1+arr2)

#SUBTRACTING.

print(arr1-arr2)

# MULTIPLICATION

print(arr1*arr2)

# DIVISION

print(arr1/arr2)

#IT THROWS ERROR WHEN THE DIMENSIONS ARE DIFFERENT.
```

```python
# MAKING SAME DIMENSIONS WITHOUT KNOWING ORIGINAL ONE.
arr = np.arange(12).reshape(3,4)
print(arr)

print(np.mean(arr))
print(np.mean(arr,axis=0)) # axis tells us which way it is going to take the mean (AT ROW LEVEL)
print(np.mean(arr,axis=1)) #AT COLOUMN LEVEL.
print(arr.ndim) #IT WILL SHOW THE DIMENSIONS/SHAPE OF TUPPLE.
```

```python
# LETS TRY MERGING.
product_data = {
    'product_id': ['p1','p2','p3','p4'],
    'product_name':['product A','product B','product C','product D']
}
product_df = pd.DataFrame(product_data)
product_df
```

```python
review_data = {
    'product_id': ['p1','p2','p3'],
    'review_score': [4.5, 3.8, 4.0],
    'review_comment': ['Great product','Need improvment','Satisfied']
}
review_df = pd.DataFrame(review_data)
review_df

```

```python
pd.merge(product_df,review_df,on='product_id',how='inner')
```

```python
product_df.merge(review_df,on='product_id',how='inner')
```

```python
pd.merge(product_df,review_df,on='product_id',how='left')
```

```python
pd.merge(product_df,review_df,on='product_id',how='right')
```
